```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 128x128 image with random black and white pixels
image_size = 128
num_pixels = image_size * image_size // 2

# Generate random indices for black pixels
black_indices = np.random.choice(image_size * image_size, num_pixels, replace=False)

# Create an empty image
image = np.zeros((image_size, image_size), dtype=np.uint8)

# Set black pixels
image.flat[black_indices] = 255

# Display the image
plt.imshow(image, cmap='gray')
plt.axis('off')  # Hide axes
plt.show()
```
